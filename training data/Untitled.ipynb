```python
import json
import os
from tqdm import tqdm

# =========================
# USER INPUT
# =========================

# Folder containing your JSON files
folder_path = "./"   # change if needed

# Output file
output_file = "combined_output.json"

# =========================
# GET ALL JSON FILES
# =========================

file_list = [
    os.path.join(folder_path, f)
    for f in os.listdir(folder_path)
    if f.endswith(".json")
]

print(f"Found {len(file_list)} JSON files.\n")

# =========================
# PROCESSING
# =========================

combined_data = []

for file_path in tqdm(file_list, desc="Processing files"):
    try:
        with open(file_path, 'r') as f:
            data = json.load(f)

        for obs in data:
            # Add missing field
            if "vehicle_load_capacity" not in obs:
                obs["vehicle_load_capacity"] = 50

            # Remove unwanted fields
            obs.pop("instance_id", None)
            obs.pop("nodes_execluded", None)

            # Keep only required fields
            cleaned_obs = {
                "experiment_id": obs.get("experiment_id"),
                "instance_type": obs.get("instance_type"),
                "instance_index": obs.get("instance_index"),
                "objective_value": obs.get("objective_value"),
                "customers_not_selected_first_place": obs.get("customers_not_selected_first_place"),
                "customers_not_visited_from_selected_pool": obs.get("customers_not_visited_from_selected_pool"),
                "battery_capacity": obs.get("battery_capacity"),
                "vehicle_load_capacity": obs.get("vehicle_load_capacity"),
                "feasibility_prize": obs.get("feasibility_prize"),
                "best solution": obs.get("best solution"),
            }

            combined_data.append(cleaned_obs)

    except Exception as e:
        print(f"\n❌ Error processing {file_path}: {e}")

# =========================
# SAVE OUTPUT
# =========================

with open(output_file, 'w') as f:
    json.dump(combined_data, f, indent=4)

print(f"\n✅ Done! Combined file saved as: {output_file}")
print(f"Total records: {len(combined_data)}")
```
